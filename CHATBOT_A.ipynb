```python
import PyPDF2
import requests
from bs4 import BeautifulSoup
import re

def read_pdf(file_path):
    text = ""
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            extracted_text = page.extract_text()
            if extracted_text:
                text += extracted_text + "\n"
    return text.lower()

def scrape_website(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers, timeout=10)
    soup = BeautifulSoup(response.text, "html.parser")
    paragraphs = soup.find_all("p")
    text = " ".join(p.get_text() for p in paragraphs)
    return text.lower()

def extract_keywords(question):
    stopwords = {
        "how","to","is","the","a","an","of","for","in","on",
        "was","were","are","as","by","with","that","this","from"
    }
    words = re.findall(r"\b[a-zA-Z]+\b", question.lower())
    return [w for w in words if w not in stopwords]

def score_sentences(text, keywords):
    sentences = re.split(r"[.\n]", text)
    results = []

    for sentence in sentences:
        score = sum(1 for keyword in keywords if keyword in sentence)
        if score > 0:
            results.append((score, sentence.strip()))

    results = list(set(results))
    results.sort(reverse=True, key=lambda x: x[0])
    return results

def chatbot_response(question, pdf_text, web_text):
    keywords = extract_keywords(question)
    final_response = []

    web_results = score_sentences(web_text, keywords)
    if web_results:
        final_response.append("ChatBot (From Website):")
        for _, sentence in web_results[:5]:
            final_response.append("- " + sentence)

    pdf_results = score_sentences(pdf_text, keywords)
    if pdf_results:
        final_response.append("\nChatBot (From PDF):")
        for _, sentence in pdf_results[:5]:
            final_response.append("- " + sentence)

    if final_response:
        return "\n".join(final_response)

    return "Sorry, no relevant instruction found."

# Load sources
pdf_text = read_pdf("/content/CN_LAB_2330051.pdf")
web_text = scrape_website("https://en.wikipedia.org/wiki/Narendra_Modi")

print("Chatbot is ready. Type 'exit' to quit.")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Bot: Goodbye!")
        break

    response = chatbot_response(user_input, pdf_text, web_text)
    print("Bot:\n", response)
```
